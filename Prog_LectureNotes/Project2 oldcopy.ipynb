# Introduction

Ice cores, cylindrical samples drilled from ice sheets or glaciers, primarily extracted from Antarctica and Greenland, with the longest ones reaching depths of 3km. The oldest continuous ice core records, spanning 123,000 years in Greenland and 800,000 years in Antarctica, provide insights into past temperature and various environmental aspects. These ice cores contain small air bubbles that hold samples of the ancient atmosphere, enabling direct measurement of past concentrations of key greenhouse gases like carbon dioxide, methane, and nitrous oxide.[1]
Analyzing Antarctic ice cores reveals historical changes in atmospheric carbon dioxide levels. Air bubbles in Antarctic Vostok and EPICA Dome C ice cores offer a composite record of carbon dioxide levels spanning 650,000 years. More recent data from the lowest 200 meters of the Dome C core extend this record to 800,000 years, revealing the lowest carbon dioxide concentration measured in an ice core. Carbon dioxide correlates strongly with Antarctic temperature over eight glacial cycles, with notably lower concentrations between 650,000 and 750,000 years ago.[2]<br>
Understanding the onset of anthropogenic climate change is crucial for adapting to its impacts. Ireland, situated on the western European Atlantic seaboard, can be considered a sentinel location. Analysis of long-term climate data may reveal climate change signals for Ireland, highlighting by its susceptibility to climate variability.[3]<br>
Reproducible climate reconstructions of the Common Era are key to placing industrial-era warming into the context of natural climatic variability.[4] However, datasets that would combine prehistoric and common era records alike are not readily available. <br>
This project aims at gathering, combining and examining such multiproxy data, predicting future trends based on the gathered climate data and also making it available in accessible JSON and CSV format.<br>

*"The emergence of a climate change signal in long-term Irish meteorological observations"[5]* study, funded by the Irish Environmental Protection Agency and Met Éireann, examines the emergence of a forced anthropogenic climate change signal in temperature and precipitation observations for Ireland. Using regression analysis with the annual Global Mean Surface Temperature (GMST), the researchers assess the signal across various indices, categorizing the signal-to-noise ratio as normal, unusual, or unfamiliar compared to the early industrial climate. Over half of the indices indicate the emergence of at least unusual conditions. The annual mean temperature increase has resulted in unfamiliar climates at six of eleven stations. Warming at the island scale is estimated at 0.88 °C per degree of GMST warming. While many stations show unusual climates for spring, summer, and autumn mean temperatures, no forced signal is found for winter mean temperature. Changes in cool/warm days and nights are unfamiliar, but no anthropogenic signal is found for the hottest day annually or in summer. Increases in annual precipitation, especially in winter, suggest heightened flood risk with continued warming. Summer precipitation shows no significant relationship with GMST. The analysis indicates a discernible climate change signal for Ireland, a region strongly influenced by climate variability.

### 1.2.1 Generating composite temperature data 

Initial temperature dataset was made available for download by the lecturer and was imported from "EPICA Dome C Ice Core 800KYr Deuterium Data and Temperature Estimate 2007.txt" from the 'Data' folder after manual download. The 'Temperature' variable is referring to Temperature anomaly that in the context of palaeoclimatology means the deviation from a long-term average, in this case Temperature difference from the average of the last 1000 years. [1]

> ###### [1] Jouzel et al. (2007) Orbital and Millennial Antarctic Climate Variability over the Past 800,000 Years., Science, Vol. 317, No. 5839, pp.793-797, 10 August 2007.

```python
# read in temperature from original source "EPICA Dome C Ice Core 800KYr Deuterium Data and Temperature Estimate 2007.txt" 
dfTemperature = pd.read_csv('Data/EPICA Dome C Ice Core 800KYr Deuterium Data and Temperature Estimate 2007.txt', sep='\s+', skiprows=91)
dfTemperature.head()
print(dfTemperature.isnull().sum())# check for Null entries (absence of value)(ref:  https://datatofish.com/check-nan-pandas-dataframe/)
dfTemperature.describe()
#read in text data separated with multiple spaces (ref: https://stackoverflow.com/questions/19632075/how-to-read-file-with-space-separated-values-in-pandas)
```

The test for missing values returned with 12 and 15 rows in Deuterium and Temperature columns respectively.<br>
The dataframe has between 5788 and 5800 entries in 5 columns in 'int64' numeric format. According to the data description in the source file, two columns are relevant for this project **"Column 3 (Age): EDC3 age scale (years before year 1950)"** and **"Column 5(Temperature): Temperature estimate (temperature difference from the average of the last 1000 years)"**. [1] <br>
The time-scale, used in column 3 is referred to as BP or Before Present timescale, used mainly in archaeology, geology and other scientific disciplines to specify when events occurred relative to the origin of practical radiocarbon dating in the 1950s. [2] [3] <br>
In the next step, columns that are not relevant to this project are being dropped, along with rows with missing values to avoid run-time errors, then summary statistics repeated with the modified dataframe.

> ###### [1] Jouzel et al. (2007) Orbital and Millennial Antarctic Climate Variability over the Past 800,000 Years., Science, Vol. 317, No. 5839, pp.793-797, 10 August 2007., <br> [2] Wikipedia (2023), Before Present, https://en.wikipedia.org/wiki/Before_Present <br>

```python
# data cleaning - removing missing values and non relevant data
dfTemperature=dfTemperature.drop(['Bag','ztop','Deuterium'],axis=1) # drop columns ref: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html
dfTemperature=dfTemperature.dropna() # drop rows with missing values ref: https://www.w3schools.com/python/pandas/ref_df_dropna.asp
dfTemperature = dfTemperature.reset_index(drop=True) # reset index after dropping rows ref: https://note.nkmk.me/en/python-pandas-reset-index/
print(dfTemperature.isnull().sum())
dfTemperature.describe()
```

The resulting dataframe has 5785 entries without missing values. Age column has a minimum value of 38, indicating that the latest sample is from 1988 while the maximum value is over 800 thousand marking the age of the oldest sample. Temperature data shows an average -4.5 temperature difference from the average of the last 1000 years. As the original source does not contain samples from modern times, a suitable supplementary dataset was sourced. The new data is from the "PAGES2k Global 2,000 Year Multiproxy Database"[1] that has temperature data from the last 2000 years up until 2005. The new source is imported into a dataframe, column names aligned with the existing data frame and the two dataframes concatenated into one.

> ###### [1] PAGES2k Consortium. (2017) A global multiproxy database for temperature reconstructions of the Common Era. Scientific Data, 4, 170088. doi: 10.1038/sdata.2017.88 <br>

```python
# read in modern global temperature from proxy source "temperature_Arc-HalletLake.McKay.2008.txt" 
dfTemp_modern = pd.read_csv('Data/temperature_Arc-HalletLake.McKay.2008.txt', sep='\s+', skiprows=222)
dfTemp_modern = dfTemp_modern.drop(['year','uncertainty_temperatureduration','thickness','BSi'],axis=1)
dfTemp_modern=dfTemp_modern.rename(columns={'age':'Age','temperature':'Temperature'})
Temperature = pd.concat([dfTemp_modern, dfTemperature], axis=0, ignore_index=True)
Temperature.describe()
#read in text data separated with multiple spaces (ref: https://stackoverflow.com/questions/19632075/how-to-read-file-with-space-separated-values-in-pandas)
```

With the addition of the PAGES2k records, minimum age data at -55 confirms that the most recent sample is from 2005. To cover the remaining years a further data source from NOAA National Centers for Environmental information was added.[1]

In order to visualise each dataframe, a dataframe list was stored in 'dataframes' variable, the data sources and citations stored in 'labels' variable and an additional variable was created for line styling because text with the default matplotlib colour scheme were not always legible. A **for** loop iterates through all three lists. Within the for loop, a matplotlib plot function draws up a line plot for Age and CO2 data from each dataframe to the same plot with Age on the **x**-axis and CO2 on the **y**-axis. ### Augmented Dickey–Fuller (ADF) test

If a time series has no trend, exhibits constant variance over time, and has a constant autocorrelation structure over time, it is considered stationary'. One way to test whether a time series is stationary is to perform an augmented Dickey-Fuller test, which uses the following null (H0 - The time series is non-stationary) and alternative hypotheses (HA - The time series is stationary). If the p-value from the test is less than some significance level (e.g. α = .05), then we can reject the null hypothesis and conclude that the time series is stationary.[1]<br>
In Python the Augmented Dickey-Fuller unit root test is part of the statsmodels library as *statsmodels.tsa.stattools.adfuller()* [2] <br>
The adfuller() function outputs a list that contains multiple values regarding the results of the ADF test. The second figure in the output is the p-value. The p-value is a probability that measures how likely it is to observe a test statistic as extreme as the one calculated from the data, assuming that the null hypothesis is true. The smaller the p-value, the more likely that the time series is stationary, as it provides evidence against the null hypothesis.<br>

> ###### [1] Statology (2023), Augmented Dickey-Fuller Test in Python, ref: https://www.statology.org/dickey-fuller-test-python/ [2] statsmodels.org (2023), satsmodels.tsa.stattools.adfuller, https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html



The following code encapsulates the ADF test within a function *fn_adfuller* that creates a more user-friendly output of the native .adfuller() function with labels and allows for a quick ADF test for  to and a simple if 

```python
# interploated data
# create function with labelled output from Augmented Dickey-Fuller test
def fn_adfuller(var,age):
    #test: Time series data sliced before the age variable
    slice=Combined_CO2_T[Combined_CO2_T['Age'] < age]
    test=slice[var]
    adf = sts.adfuller(test)
    print(f'\nAugmented Dickey-Fuller test for: {var} - trend for the last {age} years') # get the label of the column
    print(f'p-value: {adf[1]}') #the probability of observing a test statistic (low = stationary)
    print(f'Lags Used: {adf[2]}') # Lags used to count for autocorrelation, to capture longer term patterns
    print(f'Number of Observations Used: {adf[3]}') #the lenght of the time series
    print('---')
    if adf[1]<0.05:
        print(f'p-value: {round(adf[1],4)} - The trend is likely stationary')
    else:
        print(f'p-value: {round(adf[1],4)} - The trend is likely NOT stationary')

fn_adfuller('CO2',805000)
fn_adfuller('CO2',262) #after industrial revolution
```

### 2.2 Seasonality and cyclic patterns

In time series data, seasonality is the presence of variations that occur at specific regular intervals less than a year, such as weekly, monthly, or quarterly. [1] In the context of palaeoclimatological data that represents many hundreds of thousands of years, seasonality is not a relevant feature of a time series to discuss. However, if we visualise the CO2 and Temperature data for the last 800k years, there is a visible pattern of increasing and decreasing CO2 and Temperature anomaly values repeating about every hundred thousand years or so. ### Combine CO2 and Temperature data into one dataframe.
As both CO2 and Temperature datasets have an 'Age' column it is a logical step to combine the two on the 'Age' axis. The callange in creating a combined dataframe including both CO2 and Temperature however, is the potential loss of a large amount of data if using the *pd.merge* function where the Age data doesn't match, or using the *concat* function which keeps all the data but leaves NaN entries where there is no matching 'Age' value. In order to keep all original data, the two DataFrames were fused with the *concat* function into a new DataFrame named Combined_CO2_T. After sorting the combined data by the 'Age' values, the missing values problem were addressed by using interpolation. The *interpolate* function estimates and fills in the gaps in the dataset. Interpolation is a technique in Python used to estimate unknown data points between two known data points. It is mostly used in time series data. [1] Following interpolation, the remaining 2 rows with missing values were removed using the dropna method. The resulting DataFrame, Combined_CO2_T, reflects a combined and interpolated dataset, providing one continuous time axis for both variables facilitating subsequent analyses.

> ###### [1] Jyothi Chanda, Interpolation using pandas (2022), https://www.numpyninja.com/post/interpolation-using-pandas


```python
#Combined CO2 and temperature data
Combined_CO2_T=pd.concat([PyCompCO2,Temperature]).copy()
Combined_CO2_T.reset_index(drop=True, inplace=True)
Combined_CO2_T.sort_values(['Age'], inplace=True)

Combined_CO2_T.interpolate(inplace=True) #interpolate missing data

Combined_CO2_T=Combined_CO2_T.dropna()
print(Combined_CO2_T.isna().sum())
Combined_CO2_T.reset_index(drop=True, inplace=True) # reset index
Combined_CO2_T.describe()
```

```python
#Plot source data and concatenated data side by side on dual axis

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5)) #set two sublots

#Original data
ax3 = ax1.twinx() # share x axis for ax3 and ax1
ax1.plot(PyCompCO2['Age'], PyCompCO2['CO2'], color='black', label='CO2', alpha=1)
ax1.set_ylabel('CO2 (ppmv)')
ax1.set_xlabel('Age')
ax3.plot(Temperature['Age'], Temperature['Temperature'], color='r', label='Δ C°', lw=1, linestyle=':', alpha=0.8)
ax3.set_ylabel('Δ C°')
ax3.set_ylim(-10, 10)
ax1.legend(loc='upper left')
ax3.legend(loc='upper right')
ax1.set_title('Original Data')

#Interpolated data
ax4 = ax2.twinx()
ax2.plot(Combined_CO2_T['Age'], Combined_CO2_T['CO2'], color='b', label='CO2 - Interpolated', alpha=1)
ax2.set_ylabel('CO2 (ppmv)')
ax2.set_xlabel('Age')
ax4.plot(Combined_CO2_T['Age'], Combined_CO2_T['Temperature'], color='g', label='Δ C° - Interpolated', lw=1, linestyle=':', alpha=0.8)
ax4.set_ylabel('Δ C°')
ax4.set_ylim(-10, 10)
ax2.legend(loc='upper left')
ax4.legend(loc='upper right')
ax2.set_title('Concatenated Data after Interpolatation')

plt.suptitle("The temperature-CO2 connection - CO2 and Temperature anomaly over 800k years")
plt.grid(True, linestyle=':', color='#ADD8E6')
plt.tight_layout()
plt.show()
```

# Time series analysis

Trend
A trend is observed when there is an increasing or decreasing slope observed in the time series.
+ increasing CO2 and temperature

Seasonality - The seasonality component exhibits a trend that repeats with respect to timing, direction, and magnitude. Some examples include an increase in water consumption in summer due to hot weather conditions.
+ winter/summer - in such long timeline it's not visible

Cyclic Variations./ Cyclical Component - These are the trends with no set repetition over a particular period of time. A cycle refers to the period of ups and downs, booms and slums of a time series, mostly observed in business cycles. These cycles do not exhibit a seasonal variation but generally occur over a time period of 3 to 12 years depending on the nature of the time series.
+ ice ages

Irregular variation - Irregular Variation - These are the fluctuations in the time series data which become evident when trend and cyclical variations are removed. These variations are unpredictable, erratic, and may or may not be random.

ETS Decomposition - ETS Decomposition is used to separate different components of a time series. The term ETS stands for Error, Trend and Seasonality.


1 Dependence- It refers to the association of two observations of the same variable at prior time periods.

2 Stationarity- It shows the mean value of the series that remains constant over the time period. If past effects accumulate and the values increase towards infinity then stationarity is not met.

3 Differencing- Differencing is used to make the series stationary and to control the auto-correlations. There may be some cases in time series analyses where we do not require differencing and over-differenced series can produce wrong estimates.

4 Specification - It may involve the testing of the linear or non-linear relationships of dependent variables by using time series models such as ARIMA models.

5 Exponential Smoothing - Exponential smoothing in time series analysis predicts the one next period value based on the past and current value. It involves averaging of data such that the non-systematic components of each individual case or observation cancel out each other. The exponential smoothing method is used to predict the short term prediction.

6 Curve fitting - Curve fitting regression in time series analysis is used when data is in a non-linear relationship.

7 ARIMA - ARIMA stands for Auto Regressive Integrated Moving Average.
