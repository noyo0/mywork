#### Seeds

a "seed" is a starting point or initial value used to initialize a random number generator (RNG). Random number generators in computer programs are typically pseudorandom, which means they generate numbers that appear random but are actually generated using a deterministic algorithm. The seed is a value that helps set the initial state of the RNG, and from this starting point, the RNG can produce a sequence of seemingly random numbers.

Seed is useful when we want to repeat the same set of random numbers (when debugging for example) 

```python
rng = np.random.default_rng(seed=42)
x=rng.random((5,3))
print(x)
```
