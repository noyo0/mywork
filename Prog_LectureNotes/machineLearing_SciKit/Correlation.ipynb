+ Negative correlation (red dots): In the plot on the left, the y values tend to decrease as the x values increase. This shows strong negative correlation, which occurs when large values of one feature correspond to small values of the other, and vice versa.

+ Weak or no correlation (green dots): The plot in the middle shows no obvious trend. This is a form of weak correlation, which occurs when an association between two features is not obvious or is hardly observable.

+ Positive correlation (blue dots): In the plot on the right, the y values tend to increase as the x values increase. This illustrates strong positive correlation, which occurs when large values of one feature correspond to large values of the other, and vice versa.

```python
#import modules

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

### Data creation
create random data for x and y axes. Have y data be a straighte line function of the x data with some noise added

```python
#equation for straight line: y=mx+b
"""
y is the dependent variable (the vertical axis),
x is the independent variable (the horizontal axis),
m is the slope of the line (rise over run),
b is the y-intercept, representing the point where the line crosses the y-axis."""

rng =np.random.default_rng() #initiate random
x=rng.random(10)*10

y=3*x+4 +(rng.random(10)*20) #equation of a line plus noise
```

#### Visualise data

```python
plt.scatter(x,y)
plt.show()
```

#### calculate correlation

```python
r=np.corrcoef(x,y) #correlation coefficient
r
```

##### energyprices

Model energyprices with random figures

```python
timeline = np.linspace(0,10) #linspace() is a function in the NumPy library in Python that is used to create evenly spaced values over a specified range. 
energyPrice = np.sin(timeline) #sinus will give a sinus curve
plt.plot(timeline,energyPrice,'k')
sharePrice = (np.sin(timeline)*3)+((rng.random(50)*2)-1)#added noise
plt.plot(timeline,sharePrice,'r')
plt.show()
```

```python
#check correlation
r=np.corrcoef(sharePrice,energyPrice)
r
```
