# Introduction

Ice cores, cylindrical samples drilled from ice sheets or glaciers, primarily extracted from Antarctica and Greenland, with the longest ones reaching depths of 3km. The oldest continuous ice core records, spanning 123,000 years in Greenland and 800,000 years in Antarctica, provide insights into past temperature and various environmental aspects. These ice cores contain small air bubbles that hold samples of the ancient atmosphere, enabling direct measurement of past concentrations of key greenhouse gases like carbon dioxide, methane, and nitrous oxide.[1]
Analyzing Antarctic ice cores reveals historical changes in atmospheric carbon dioxide levels. Air bubbles in Antarctic Vostok and EPICA Dome C ice cores offer a composite record of carbon dioxide levels spanning 650,000 years. More recent data from the lowest 200 meters of the Dome C core extend this record to 800,000 years, revealing the lowest carbon dioxide concentration measured in an ice core. Carbon dioxide correlates strongly with Antarctic temperature over eight glacial cycles, with notably lower concentrations between 650,000 and 750,000 years ago.[2]<br>
Understanding the onset of anthropogenic climate change is crucial for adapting to its impacts. Ireland, situated on the western European Atlantic seaboard, can be considered a sentinel location. Analysis of long-term climate data may reveal climate change signals for Ireland, highlighting by its susceptibility to climate variability.[3]<br>
Reproducible climate reconstructions of the Common Era are key to placing industrial-era warming into the context of natural climatic variability.[4] However, datasets that would combine prehistoric and common era records alike are not readily available. <br>
This project aims at gathering, combining and examining such multiproxy data, predicting future trends based on the gathered climate data and also making it available in accessible JSON and CSV format.<br>

*"The emergence of a climate change signal in long-term Irish meteorological observations"[3]* study, funded by the Irish Environmental Protection Agency and Met Éireann, examines the emergence of a forced anthropogenic climate change signal in temperature and precipitation observations for Ireland. Using regression analysis with the annual Global Mean Surface Temperature (GMST), the researchers assess the signal across various indices, categorizing the signal-to-noise ratio as normal, unusual, or unfamiliar compared to the early industrial climate. Over half of the indices indicate the emergence of at least unusual conditions. The annual mean temperature increase has resulted in unfamiliar climates at six of eleven stations. Warming at the island scale is estimated at 0.88 °C per degree of GMST warming. While many stations show unusual climates for spring, summer, and autumn mean temperatures, no forced signal is found for winter mean temperature. Changes in cool/warm days and nights are unfamiliar, but no anthropogenic signal is found for the hottest day annually or in summer. Increases in annual precipitation, especially in winter, suggest heightened flood risk with continued warming. Summer precipitation shows no significant relationship with GMST. The analysis indicates a discernible climate change signal for Ireland, a region strongly influenced by climate variability.

### 1.2.1 Generating composite temperature data 

Initial temperature dataset was made available for download by the lecturer and was imported from "EPICA Dome C Ice Core 800KYr Deuterium Data and Temperature Estimate 2007.txt" from the 'Data' folder after manual download. The 'Temperature' variable is referring to Temperature anomaly that in the context of palaeoclimatology means the deviation from a long-term average, in this case Temperature difference from the average of the last 1000 years. [1]

> ###### [1] Jouzel et al. (2007) Orbital and Millennial Antarctic Climate Variability over the Past 800,000 Years., Science, Vol. 317, No. 5839, pp.793-797, 10 August 2007.

```python
# read in temperature from original source "EPICA Dome C Ice Core 800KYr Deuterium Data and Temperature Estimate 2007.txt" 
dfTemperature = pd.read_csv('Data/EPICA Dome C Ice Core 800KYr Deuterium Data and Temperature Estimate 2007.txt', sep='\s+', skiprows=91)
dfTemperature.head()
print(dfTemperature.isnull().sum())# check for Null entries (absence of value)(ref:  https://datatofish.com/check-nan-pandas-dataframe/)
dfTemperature.describe()
#read in text data separated with multiple spaces (ref: https://stackoverflow.com/questions/19632075/how-to-read-file-with-space-separated-values-in-pandas)
```

The test for missing values returned with 12 and 15 rows in Deuterium and Temperature columns respectively.<br>
The dataframe has between 5788 and 5800 entries in 5 columns in 'int64' numeric format. According to the data description in the source file, two columns are relevant for this project **"Column 3 (Age): EDC3 age scale (years before year 1950)"** and **"Column 5(Temperature): Temperature estimate (temperature difference from the average of the last 1000 years)"**. [1] <br>
The time-scale, used in column 3 is referred to as BP or Before Present timescale, used mainly in archaeology, geology and other scientific disciplines to specify when events occurred relative to the origin of practical radiocarbon dating in the 1950s. [2] [3] <br>
In the next step, columns that are not relevant to this project are being dropped, along with rows with missing values to avoid run-time errors, then summary statistics repeated with the modified dataframe.

> ###### [1] Jouzel et al. (2007) Orbital and Millennial Antarctic Climate Variability over the Past 800,000 Years., Science, Vol. 317, No. 5839, pp.793-797, 10 August 2007., <br> [2] Wikipedia (2023), Before Present, https://en.wikipedia.org/wiki/Before_Present <br>

```python
# data cleaning - removing missing values and non relevant data
dfTemperature=dfTemperature.drop(['Bag','ztop','Deuterium'],axis=1) # drop columns ref: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html
dfTemperature=dfTemperature.dropna() # drop rows with missing values ref: https://www.w3schools.com/python/pandas/ref_df_dropna.asp
dfTemperature = dfTemperature.reset_index(drop=True) # reset index after dropping rows ref: https://note.nkmk.me/en/python-pandas-reset-index/
print(dfTemperature.isnull().sum())
dfTemperature.describe()
```

The resulting dataframe has 5785 entries without missing values. Age column has a minimum value of 38, indicating that the latest sample is from 1988 while the maximum value is over 800 thousand marking the age of the oldest sample. Temperature data shows an average -4.5 temperature difference from the average of the last 1000 years. As the original source does not contain samples from modern times, a suitable supplementary dataset was sourced. The new data is from the "PAGES2k Global 2,000 Year Multiproxy Database"[1] that has temperature data from the last 2000 years up until 2005. The new source is imported into a dataframe, column names aligned with the existing data frame and the two dataframes concatenated into one.

> ###### [1] PAGES2k Consortium. (2017) A global multiproxy database for temperature reconstructions of the Common Era. Scientific Data, 4, 170088. doi: 10.1038/sdata.2017.88 <br>

```python
# read in modern global temperature from proxy source "temperature_Arc-HalletLake.McKay.2008.txt" 
dfTemp_modern = pd.read_csv('Data/temperature_Arc-HalletLake.McKay.2008.txt', sep='\s+', skiprows=222)
dfTemp_modern = dfTemp_modern.drop(['year','uncertainty_temperatureduration','thickness','BSi'],axis=1)
dfTemp_modern=dfTemp_modern.rename(columns={'age':'Age','temperature':'Temperature'})
Temperature = pd.concat([dfTemp_modern, dfTemperature], axis=0, ignore_index=True)
Temperature.describe()
#read in text data separated with multiple spaces (ref: https://stackoverflow.com/questions/19632075/how-to-read-file-with-space-separated-values-in-pandas)
```

```python
Temperature
```

With the addition of the PAGES2k records, minimum age data at -55 confirms that the most recent sample is from 2005. To cover the remaining years a further data source from NOAA National Centers for Environmental information was added.[1]

>###### [1] NOAA National Centers for Environmental information, Climate at a Glance: Global Time Series, published December 2023, retrieved on December 25, 2023 from https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/global/time-series

```python
#Modern temperature data from 'Global Land and Ocean Temperature Anomalies - Time Series'
#NOAA National Centers for Environmental information, Climate at a Glance: Global Time Series, published December 2023, https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/global/time-series
NOAA_Temp = pd.read_csv('Data/Global Land and Ocean 1880 - 2023 Temperature anomalies.csv', skiprows=4)
NOAA_Temp['Age']=1950-NOAA_Temp['Year'] # calculate BP age
NOAA_Temp['Temperature']=NOAA_Temp['Anomaly'] # 
NOAA_Temp = NOAA_Temp.drop(['Year','Anomaly'],axis=1) #drop columns
Temperature = pd.concat([Temperature,NOAA_Temp], axis=0, ignore_index=True).sort_values('Age')
Temperature.reset_index(drop=True, inplace=True) #reset index after concat ref: https://stackoverflow.com/questions/35528119/pandas-recalculate-index-after-a-concatenation
Temperature.describe()
```

Minimum age at -73 confirms up-to-date Temperature data in the dataset that is now contains over 6000 entries. To confirm data distribution, a set of histograms are rendered for both 'Age' and 'Temperature' variables. 

```python
# Histogram for temperature data
fig, axs = plt.subplots(1,2, figsize=(10,3)) # 1 row, 3 columns
# Age
axs[0].hist(Temperature['Age'],label="Age",bins=50)
axs[0].legend()
# Temperature
axs[1].hist(Temperature['Temperature'],label="Temperature", bins=50)
axs[1].legend()
plt.suptitle("Temperature data distribution")
plt.show()
```

The histograms suggest that the majority of the samples are from the more recent periods and older samples are much more rare than modern ones, however, there are no significant gaps in the age distribution of the data which allows for a more accurate and comprehensive analysis. The temperature histogram indicates a more or less bi-modal distribution with peaks at -7.5 and 0.0 degrees. ### Augmented Dickey–Fuller (ADF) test

If a time series has no trend, exhibits constant variance over time, and has a constant autocorrelation structure over time, it is considered stationary'. One way to test whether a time series is stationary is to perform an augmented Dickey-Fuller test, which uses the following null (H0 - The time series is non-stationary) and alternative hypotheses (HA - The time series is stationary). If the p-value from the test is less than some significance level (e.g. α = .05), then we can reject the null hypothesis and conclude that the time series is stationary.[1]<br>
In Python the Augmented Dickey-Fuller unit root test is part of the statsmodels library as *statsmodels.tsa.stattools.adfuller()* [2] <br>
The adfuller() function outputs a list that contains multiple values regarding the results of the ADF test. The second figure in the output is the p-value. The p-value is a probability that measures how likely it is to observe a test statistic as extreme as the one calculated from the data, assuming that the null hypothesis is true. The smaller the p-value, the more likely that the time series is stationary, as it provides evidence against the null hypothesis.<br>

> ###### [1] Statology (2023), Augmented Dickey-Fuller Test in Python, ref: https://www.statology.org/dickey-fuller-test-python/ [2] statsmodels.org (2023), satsmodels.tsa.stattools.adfuller, https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html

The following code encapsulates the ADF test within a function *fn_adfuller* that creates a more user-friendly output of the native .adfuller() function with labels [1] and a categorization algorithm. <br> The ADF will compare CO2 data for the complete time series with samples since the industrial revolution (263 years ago) where the trend is clearly visible.

>###### kaggle.com (2019), Time Series Prediction Tutorial with EDA, https://www.kaggle.com/code/kanncaa1/time-series-prediction-tutorial-with-eda



```python
# create function with labelled output from Augmented Dickey-Fuller test
def fn_adfuller(var,age):
    #test: Time series data sliced before the age variable
    slice=Combined_CO2_T[Combined_CO2_T['Age'] < age]
    test=slice[var]
    adf = sts.adfuller(test)
    print(f'\nADF test for: {var} - trend for the last {age} years') # get the label of the column
    print(f'p-value: {adf[1]}') #the probability of observing a test statistic (low = stationary)
    print(f'Lags Used: {adf[2]}') # Lags used to count for autocorrelation, to capture longer term patterns
    print(f'Number of Observations Used: {adf[3]}') #the lenght of the time series
    if adf[1]<0.05:
        print('-The trend is likely stationary-')
    else:
        print('-The trend is likely NOT stationary-')

fn_adfuller('CO2',805000)
fn_adfuller('CO2',273) #after industrial revolution
```

The result is unexpected. According to the visualisation there is an upwards trend towards modern times. The ADF function is repeated with the original data to rule out possible effects of the interpolation.

```python
def fn_adfuller2(TS,name=''):
    #TS: Time series data
    adf = sts.adfuller(TS)
    print(f'\nADF test for:{TS.name} - {name}') # get the name of the column
    print(f'p-value: {adf[1]}') #the probability of observing a test statistic (low = stationary)
    if adf[1]<0.05:
        print('--Time sersies is likely stationary--')
    else:
        print('--Time sersies is likely NOT stationary--')

#Full data
fn_adfuller2(Temperature['Temperature'], '800k years')
fn_adfuller2(PyCompCO2['CO2'],'800k years')

#after industrial revolution 1760
PyCompCO2_af = PyCompCO2[PyCompCO2['Age'] < 273].copy() #use copy to avoid error "A value is trying to be set on a copy of a slice from a DataFrame."
Temperature_af = Temperature[Temperature['Age'] < 273].copy()
fn_adfuller2(Temperature_af['Temperature'], 'after Industrial Revolution')
fn_adfuller2(PyCompCO2_af['CO2'],'after Industrial Revolution')
```

### 2.2 Seasonality and cyclic patterns

In time series data, seasonality is the presence of variations that occur at specific regular intervals less than a year, such as weekly, monthly, or quarterly. [1] In the context of palaeoclimatological data that represents many hundreds of thousands of years, seasonality is not a relevant feature of a time series to discuss. Repeated fluctuations around the trend that are longer in duration than irregularities but shorter than what would constitute a trend are considered cycles.[2]

### Autocorrelation Function to detect randomness or patterns

Autocorrelation (APF) is a mathematical representation of the degree of similarity between a given time series and a lagged version of itself over successive time intervals. [1]<br>
Autocorrelation plots are a commonly-used tool for checking randomness in a data set. This randomness is ascertained by computing autocorrelations for data values at varying time lags. If random, such autocorrelations should be near zero for any and all time-lag separations. If non-random, then one or more of the autocorrelations will be significantly non-zero.[2]<br>
APF is the correlation between the observation at the current time spot and the observations at previous time spots. The autocorrelation function starts a lag 0, which is the correlation of the time series with itself and therefore it is always 1. [2]
Partial Autocorrelation (PACF) is estimated by controlling the effects of other lags, generally using a linear regression. The PACF at a given lag is the coefficient of that lag obtained from linear regression.[3]
For the visualisation of the ACF and PACF, *statsmodels.graphics.tsaplots* plot_acf() and plot_pacf() from the statsmodels library is utilised.

> ###### [1] (https://www.investopedia.com/terms/a/autocorrelation.asp) <br> [2] Box, G. E. P., and Jenkins, G. (1976), Time Series Analysis: Forecasting and Control, Holden-Day., https://www.itl.nist.gov/div898/handbook/eda/section4/eda43.htm#Box <br>[3] (https://www.kaggle.com/code/iamleonie/time-series-interpreting-acf-and-pacf/notebook?scriptVersionId=80259378) <br> [4] SPUR Economics (2023), Interpreting ACF and PACF plots, https://spureconomics.com/interpreting-acf-and-pacf-plots/

```python
from pandas.plotting import autocorrelation_plot
fig, axs = plt.subplots(1,2, figsize=(16,5)) # 1 row, 2 columns
plt.suptitle('Autocorrelation test to detect cycles or trends')
autocorrelation_plot(fn_slice()['CO2'], ax=axs[0])
autocorrelation_plot(fn_slice()['Temperature'], ax=axs[1])
axs[0].set_title('CO2 Autocorrelation')
axs[1].set_title('Temperature Autocorrelation')
plt.show()
```

```python
# modified ADF function for CE dataset
def fn_adfullerCE(var,yr=1750):
    #test: Time series data sliced before the age variable
    slice=CE_combined[CE_combined['Year'] > yr]
    test=slice[var]
    adf = sts.adfuller(test)
    print(f'\nADF test for: {var} - since {yr}CE') # get the label of the column
    print(f'p-value: {adf[1]}, Lags Used: {adf[2]}, Number of Observations Used: {adf[3]}') #the probability of observing a test statistic (low = stationary)
    if adf[1]<0.05:
        print('-The trend is likely stationary-')
    else:
        print('-The trend is likely NOT stationary-')
```

```python
#modified ADF test for Irish Rainfall data allowing slicing (default at start value @1850)
def fn_adfullerIRF(var,yr=1850):
    #test: Time series data sliced before the age variable
    slice=IrishRainfall[IrishRainfall.index.values > yr]
    test=slice[var]
    adf = sts.adfuller(test)
    print(f'\nADF test for: {var} - since {yr}CE') # get the label of the column
    print(f'p-value: {adf[1]}, Lags Used: {adf[2]}, Number of Observations Used: {adf[3]}') #the probability of observing a test statistic (low = stationary)
    if adf[1]<0.05:
        print('-The trend is likely stationary-')
    else:
        print('-The trend is likely NOT stationary-')
```

```python
#Augmented Dickey-Fuller test for trend detextion
fn_adfullerIRF('Avg')
fn_adfullerIRF('Avg',1960)
```

Autocorrelation plots (Box and Jenkins, pp. 28-32) are a commonly-used tool for checking randomness in a data set. This randomness is ascertained by computing autocorrelations for data values at varying time lags. If random, such autocorrelations should be near zero for any and all time-lag separations. If non-random, then one or more of the autocorrelations will be significantly non-zero.[1]
