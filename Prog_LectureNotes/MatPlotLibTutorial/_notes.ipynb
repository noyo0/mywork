## Plot two graphs on one axle

```python
#Import numpy as well
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot 2 calculated variables
x = np.arange(0.0,10.0,0.01)
y = 3.0 *x

plt.plot(x, #data for x axle
         y,#, data for y axle, 
         'ro')#bo is for blue dots instead of line ('ro' is for red dots)
plt.ylabel('other numbers')
plt.show()
```

```python
#plot 3 calculated variables; x, y and noise
x = np.arange(0.0,10.0,0.01)
y = 3.0 *x
noise = np.random.normal(0.0,1.0,(len(x)))

#first plot same as before
plt.plot(x, #data for x axle
         y,#, data for y axle, 
         'g-')#bo is for blue dots instead of line ('ro' is for red dots)
#second plot includes the new 'noise' variable
plt.plot(x, #data for x axle
         noise,#, data for y axle, 
         'r')# line design
plt.ylabel('other numbers')
plt.show()
```

```python
#plot 3 calculated variables; x, y and noise, with combining variables y+noise
x = np.arange(0.0,10.0,0.01)
y = 3.0 *x
noise = np.random.normal(0.0,1.0,(len(x)))

#first plot same as before
plt.plot(x, #data for x axle
         y,#, data for y axle, 
         'g-')#bo is for blue dots instead of line ('ro' is for red dots)
#second plot includes the new 'noise' variable
plt.plot(x, #data for x axle
         y + noise,#, data for y axle has 'y' and 'noise' together, 
         'p:')# line design
plt.ylabel('other numbers')
plt.show()
```

## Titles, legends and labels video

```python
#adding title etc.
#Order of the graphs determine what is beind drawn first. 
#If data is covered with other overlapping data, it can be fixed by changing which graph to draw first
x = np.arange(0.0,10.0,0.01)
y = 3.0 *x
noise = np.random.normal(0.0,1.0,(len(x)))

#first plot same as before
plt.plot(x, #data for x axle
         y + noise,#, data for y axle has 'y' and 'noise' together, 
         'grey', # line design
         label="Actual Data") # data label
#second plot includes the new 'noise' variable
plt.plot(x, #data for x axle
         y,#, data for y axle, 
         'r',
         label="Fitted Model")

# Titles and labels
plt.title('Average speed vs. Distance covered over time') 
plt.xlabel('Average Speed (km/h)')
plt.ylabel('Distance covered(km)')
plt.show()
```

## Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
#plot standard deviation random data on line graph
x = np.random.normal(0.0, 1.0, 1000)
plt.plot(x)
```

```python
#plot standard deviation random data on histogram
x = np.random.normal(1.0, 10.0, 1000)
plt.hist(x,80) #first item is data, second is the number of bins
```

## Graphs side by side (subplots)


```python
import matplotlib.pyplot as plt
import numpy as np
#plot standard deviation random data on line graph
x = np.random.normal(0.0, 1.0, 1000)
#subplot 1
plt.subplot(2,1,1)#number of rows, number of columns, where is this graph go
plt.plot(x)
#subplot2
plt.subplot(2,1,2)#number of rows, number of columns, where is this graph go
plt.hist(x)
```

## Other plots

```python
# multi dimensional scatter plots 
import matplotlib.pyplot as plt
import numpy as np

x1 = np.random.uniform(0.0,10.0,100)
x2 = np.random.uniform(0.0,5.0,100)
x3 = np.random.randint(0.0,20,100)
x4 = np.random.normal(0.0,50.0,100)

# x axle, y axle, colours, size of dots, transparency
plt.scatter(x1,# x axle
            x2,# y axle
            c=x3, # colours
            s=x4, # size of dots
            alpha=0.5) #transparency



```

### Powers plot

```python
#powers plot
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1.0, 15.0, 0.1)
plt.plot(x, x**2, 'g-',label='x^2')
plt.plot(x, x**3, 'b-',label='x^3')
plt.plot(x, x**4, 'r-',label='x^4')
plt.plot(x, 2**x, 'k-',label='2^x')

plt.legend()
```
